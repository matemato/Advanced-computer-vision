```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from ex2_utils import get_patch
```

```python
im = cv2.imread("./Sequences/ball/00000000.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)/255
pos = np.array([223,137])
size = np.array([55,55])
p,_ = get_patch(im, pos, size)
plt.imshow(p)
plt.show()
```
