```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from ex2_utils import get_patch
```

```python
im = cv2.imread("./Sequences/ball/00000000.jpg")
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)/255
pos = np.array([223,137])
size = np.array([55,55])
p,_ = get_patch(im, pos, size)
plt.imshow(p)
plt.show()
```

```python
rotate_matrix = cv2.getRotationMatrix2D(center=size/2, angle=3, scale=0.95)
  
# rotate the image using cv2.warpAffine 
# 90 degree anticlockwise
rotated_image = cv2.warpAffine(
    src=p, M=rotate_matrix, dsize=size)
plt.imshow(rotated_image)
plt.show()
```

```python
p,_ = get_patch(im, pos, size)
log_im = np.log(p+1)
log_im = (log_im - np.mean(log_im))
print(np.max(log_im))
print(np.min(log_im))
plt.imshow(log_im)
plt.show()
log_im /= np.std(log_im)
print(np.max(log_im))
print(np.min(log_im))
plt.imshow(log_im)
plt.show()
p,_ = get_patch(im, pos, size)
plt.imshow(p)
plt.show()
```
